```python
Q10. How can you tell the difference between break and continue?
Ans . The break statement will move the execution outside and just after a loop.
The continue statement will move the execution to the start of the loop.
```

```python
# use of break
for i in range(10):
    if(i==7):
        break
    print(i)
    
print('Breaked')
#use of  continue
for i in range(10):
    if(i==7):
        continue
    print(i)
```

```python
11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

Ans. From the above output we can conclude that they all do the same thing.
The range(10) call ranges from 0 up to (but not including) 10, range(0, 10) tells the loop to start at 0,
and range(0, 10, 1) tells the loop to increase the variable by 1 on each iteration.
```

```python
for i in range(10):
    print(i)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
for i in range(0,10):
    print(i)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
for i in range(0,10,1):
    print(i)
```
